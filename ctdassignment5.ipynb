```python
#7
import glob

files = glob.glob("/kaggle/input/code-the-dream-assignment-6/*.csv")

df_all = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

df_all.info()
```

```python
name_counts = df_all["Name"].value_counts()
good_names = list(name_counts[name_counts > 2].index)

df_all["Name"] = df_all["Name"].apply(
    lambda x: x if x in good_names else process.extractOne(x, good_names)[0]
)
```

```python
address_counts = df_all["Address"].value_counts()
good_addresses = list(address_counts[address_counts > 2].index)

df_all["Address"] = df_all["Address"].apply(
    lambda x: x if x in good_addresses else process.extractOne(x, good_addresses)[0]
)
```

```python
def fix_mode(group):
    group = group.dropna()
    return group.mode().iloc[0] if not group.mode().empty else np.nan

df_all["Zip"] = df_all.groupby(["Name", "Address"])["Zip"].transform(fix_mode)
df_all["Phone"] = df_all.groupby(["Name", "Address"])["Phone"].transform(fix_mode)
```

```python
df_all = df_all.drop_duplicates()

df_all.info()
```
